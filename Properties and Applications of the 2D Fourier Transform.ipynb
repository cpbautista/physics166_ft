# Activity 1. Anamorphic Property of FT of different 2D patterns

```python
import numpy as np
from numpy.fft import ifft2, fft2,fftshift
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.pyplot import imshow, imread
import random
```

1. Create the following patterns, take their Fourier Transform and display their FT modulus. You may either generate them in Scilab for better symmetry or hand draw them in Paint.

    a. Tall rectangle aperture
    
    b. Wide rectangle aperture
    
    c. Two dots along x-axis symmetric about the center 
    
    d. Same as (c) with different spacing between the dots.

```python
size = 128
tall_aperture = np.zeros((size,size))

for i in range(size):
    for j in range(size):
        if i in range(size) and j in range(59,69):
            tall_aperture[i][j] = 1

plt.figure(figsize=(9,4.5))
plt.subplot(1, 2, 1)
imshow(tall_aperture)

plt.subplot(1, 2, 2)
imshow(fftshift(abs(fft2(tall_aperture))))

plt.tight_layout()
plt.show()
```

```python
size = 128
wide_aperture = np.zeros((size,size))

for i in range(size):
    for j in range(size):
        if j in range(size) and i in range(59,69):
            wide_aperture[i][j] = 1
            
plt.figure(figsize=(9,4.5))
plt.subplot(1, 2, 1)
imshow(wide_aperture)

plt.subplot(1, 2, 2)
imshow(fftshift(abs(fft2(wide_aperture))))

plt.tight_layout()
plt.show()
```

```python
def make_circle(ratio,distance):
    circle_size = 256
    circle = np.zeros((circle_size,circle_size))

    for i in range(circle_size):
        for j in range(circle_size):
            if np.sqrt((circle_size/2 - i)**2 + ((circle_size-distance)/2 - j)**2) <= (circle_size)//ratio:
                circle[i,j] = 1
    for i in range(circle_size):
        for j in range(circle_size):
            if np.sqrt((circle_size/2 - i)**2 + ((circle_size+distance)/2 - j)**2) <= (circle_size)//ratio:
                circle[i,j] = 1
    return circle

two_dots = make_circle(50,10)

plt.figure(figsize=(9,4.5))
plt.subplot(1, 2, 1)
imshow(two_dots)

plt.subplot(1, 2, 2)
imshow(fftshift(abs(fft2(two_dots))))

plt.tight_layout()
plt.show()
```

```python
two_dots = make_circle(50,30)
plt.figure(figsize=(9,4.5))
plt.subplot(1, 2, 1)
imshow(two_dots)

plt.subplot(1, 2, 2)
imshow(fftshift(abs(fft2(two_dots))))

plt.tight_layout()
plt.show()
```

5. Replace the dots with Gaussians, i.e., expxo2  of varying 2
variance 2. Here, o are the peak location. You will need to create this image in Scilab. Observe the FT modulus as you vary the variance.

```python
def gauss(var):
    peak = 60
    x1 = np.linspace(-128,128,100)
    y1 = x1
    x,y = np.meshgrid(x1,y1)
    a = np.exp(-((x-peak)**2 + y**2)/(var**2))
    b = np.exp(-((x+peak)**2 + y**2)/(var**2))
    
    return a+b
```

```python
gauss_1 = gauss(1)

plt.figure(figsize=(9,4.5))
plt.subplot(1, 2, 1)
imshow(gauss_1)

plt.subplot(1, 2, 2)
imshow(fftshift(abs(fft2(gauss_1))))

plt.tight_layout()
plt.show()
```

```python
gauss_1 = gauss(10)

plt.figure(figsize=(9,4.5))
plt.subplot(1, 2, 1)
imshow(gauss_1)

plt.subplot(1, 2, 2)
imshow(fftshift(abs(fft2(gauss_1))))

plt.tight_layout()
plt.show()
```

```python
gauss_1 = gauss(40)

plt.figure(figsize=(9,4.5))
plt.subplot(1, 2, 1)
imshow(gauss_1)

plt.subplot(1, 2, 2)
imshow(fftshift(abs(fft2(gauss_1))))

plt.tight_layout()
plt.show()
```

6. Create a 200200 array of zeros. Put 10 1's in random locations in the array. These ones will approximate dirac deltas. Call this array A. Create an arbitrary 99 pattern, call it d. Convolve A and d. What do you observe?

```python
random.seed(500)
A = np.zeros((200,200))
for i in range(10):
    j = random.randrange(200)
    k = random.randrange(200)
    A[j][k] = 1
imshow(A, cmap='gray')
```

```python
d = np.zeros((200, 200))
d_patt = rgb2gray(plt.imread('flake.png'))    
r, c = 99, 99
d[r:r+d_patt.shape[0], c:c+d_patt.shape[1]] += d_patt
plt.imshow(d, cmap = 'gray')
```

```python
FA = fft2(A)
Fd = fft2(d)
FAd = FA * np.conj(Fd)
IAd = ifft2(FAd)
FImage = abs(IAd)
imshow(FImage, cmap='gray')
```

#### The 9x9 pattern is "pasted" on each of the random 1s on the array A.

# Activity 4. Fingerprints : Ridge Enhancement

1. Prepare an image of your own fingerprint in grayscale. 

```python
plt.figure(figsize=(7,7))
imshow(imread('fprint.png'))
```

2. Open in Scilab as grayscale image.

```python
plt.figure(figsize=(7,7))
fprint = rgb2gray(plt.imread('fprint.png'))  
imshow(fprint)
```

3. Obtain the FT of the image and investigate where the frequencies of the fingerprint ridges lie.

```python
plt.figure(figsize=(7,7))
ridges = np.log(abs(fftshift(fft2(fprint))))
imshow(ridges)
```

```python
def save_image(data, cm, fn):
   
    sizes = np.shape(data)
    height = float(sizes[0])
    width = float(sizes[1])
     
    fig = plt.figure()
    fig.set_size_inches(width/height, 1, forward=False)
    ax = plt.Axes(fig, [0., 0., 1., 1.])
    ax.set_axis_off()
    fig.add_axes(ax)
 
    ax.imshow(data, cmap=cm)
    plt.savefig(fn, dpi = height) 
    plt.close()
```

```python
save_image(ridges,'gray','ridges.png')
```

4. Design a filter mask to enhance the appearance of these ridges at the same time remove the blotches.

```python
mask = rgb2gray(plt.imread('mask.png'))  
plt.figure(figsize=(7,7))
imshow(mask)
```

```python
Fvip = fft2(fprint)
Fape = fftshift(mask)
Fprod = Fvip * Fape
clean = abs(ifft2(Fprod))
plt.figure(figsize=(7,7))
plt.imshow(clean)
```

```python
plt.figure(figsize=(9,9))
plt.subplot(1, 2, 1)
plt.title("Original grayscale image")
imshow(fprint)

plt.subplot(1, 2, 2)
plt.title("Cleaned fingerprint")
imshow((clean), cmap="gray")

plt.tight_layout()
plt.show()
```
